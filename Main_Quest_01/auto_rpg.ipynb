```python
import random

# Q1
class Character:
    def __init__(self, name, level, hp, attack, defense):   # 속성 셍성
        self.name = name        # 이름
        self.level = level      # 레벨
        self.hp = hp            # 체력
        self.attack = attack    # 공격력
        self.defense = defense  # 방어력

    def is_alive(self):         # 잔여 HP 0이하 유무 (배틀 종료 기준)
        return self.hp > 0

    def take_damage(self, damage):  # 실제 데미지 메서드
        actual_damage = damage - self.defense
        if actual_damage > 0:
            self.hp -= actual_damage
        print(f"{self.name}은 {actual_damage}만큼의 데미지를 받았다.\n[남은 {self.name}의 HP: {self.hp}]\n")

    def attack_target(self, target):    # 공격 메서드
        damage = random.randint(1, self.attack)
        print(f"{self.name}이(가) {target.name}에게 {damage}만큼 공격했다...!")
        target.take_damage(damage)

# Q2
class Player(Character):    # Character 상속
    def __init__(self, name):
        super().__init__(name, level=1, hp=100, attack=25, defense=5)   # 인스턴스 생성 및 초기화
        self.experience = 0     # 경험치 속성 추가

    def gain_experience(self, exp):     # 경험치 연산 메서드
        self.experience += exp
        print(f"{self.name}이(가) {exp}의 경험치를 얻었습니다. (총 경험치: {self.experience})")

    def level_up(self):     # 레벨업 메서드
        if self.experience >= 50:
            self.level += 1
            self.attack += 10
            self.defense += 5
            self.experience -= 50
            print(f"{self.name}이(가) 레벨업! 현재 레벨: {self.level}, 공격력: {self.attack}, 방어력: {self.defense}, 체력: {self.hp}")
        
class Monster(Character):      # Character 상속
    def __init__(self, name, level):
        hp = random.randint(10, 30) * level
        attack = random.randint(5, 20) * level
        defense = random.randint(1, 5) * level
        super().__init__(name, level, hp, attack, defense)    # 몬스터 인스턴스 생성 및 속성값 랜덤 부여

# Q3
def battle(player, monster):    # 배틀 함수
    print(f"\n{monster.name}과의 전투를 시작합니다.")
    print(f"{player.name}({player.hp}) vs {monster.name}({monster.hp})\n")
    while player.is_alive() and monster.is_alive():     # Player, Monster 인스턴스의 체력이 모두 0 초과일 경우 무한 반복
        player.attack_target(monster)   # 플레이어 선공
        if not monster.is_alive():      # 몬스터 인스턴스의 체력이 0 이하일 경우 배틀 종료
            break
        monster.attack_target(player)   # 몬스터 후공
    if player.is_alive():               # 배틀 종료 후 플레이어가 살아남았을 경우
        exp = monster.level * 20
        player.gain_experience(exp)
        player.level_up()
        print("전투 승리!\n")
        return True
    else:
        print("전투 패배...\n")            # 배틀 종료 후 플레이어가 죽었을 경우
        return False

# Q4
def main():
    monster_dict = {'슬라임': 1, '고블린': 2, '오크': 3}
    player_name = input("플레이어의 이름을 입력하세요: ")
    player = Player(player_name)
    for m_name, m_level in monster_dict.items():
        monster = Monster(m_name, m_level)
        result = battle(player, monster)
        if not result:
            print("---게임 오버---")
            return
    print("***게임 클리어***")
```

```python
main()
```
